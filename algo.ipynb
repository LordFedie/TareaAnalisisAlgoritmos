```python
import pandas as pd
import matplotlib.pyplot as plt

Resultado1 = pd.read_csv("ResultadoDividePorConquistarMejorado.csv") 
Resultado2 = pd.read_csv("ResultadoFuerzaBrutaMejorado.csv")

plt.grid()
plt.plot(Resultado1["n"],Resultado1["t_mean"],label="DPCM")
plt.plot(Resultado2["n"],Resultado2["t_mean"],label="FBM")
plt.legend()
plt.xlabel("Cantidad de Puntos")
plt.ylabel("Tiempo en Nanosegundos")
plt.title("Comparaci√≥n Mejorada")
```
