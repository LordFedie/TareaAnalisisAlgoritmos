```python
import pandas as pd
import matplotlib.pyplot as plt

Resultado1 = pd.read_csv("ResultadoFuerzaBruta.csv")
Resultado2 = pd.read_csv("ResultadoDividePorConquistar.csv")
Resultado3 = pd.read_csv("ResultadoFuerzaBrutaMejorado.csv")
Resultado4 = pd.read_csv("ResultadoDividePorConquistarMejorado.csv") 




plt.grid()
plt.plot(Resultado1["n"],Resultado1["t_mean"],label="DPCM")
plt.plot(Resultado2["n"],Resultado2["t_mean"],label="FBM")
plt.plot(Resultado3["n"],Resultado3["t_mean"],label="FB")
plt.plot(Resultado4["n"],Resultado4["t_mean"],label="DPC")
plt.legend()
plt.xlabel("Cantidad de Puntos")
plt.ylabel("Tiempo en Nanosegundos")
plt.title("Comparaci√≥n De Implementaciones")
```
