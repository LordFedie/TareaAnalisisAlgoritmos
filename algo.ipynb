```python
import pandas as pd
import matplotlib.pyplot as plt

Resultado1 = pd.read_csv("ResultadoDividePorConquistar.csv") 
Resultado2 = pd.read_csv("ResultadoFuerzaBruta.csv")

plt.grid()
plt.plot(Resultado1["n"],Resultado1["t_mean"],label="DPC")
plt.plot(Resultado2["n"],Resultado2["t_mean"],label="FB")
plt.legend()
plt.xlabel("Cantidad de Puntos")
plt.ylabel("Tiempo en Nanosegundos")
plt.title("Divide Por Conquistar")
```
